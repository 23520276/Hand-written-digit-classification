```python
import os
import shutil
from tqdm import tqdm

source_dir = r"D:\code_things\do an cuoi ki mon may hoc\hand_Written(CNN)\data"
dest_dir = r"D:\code_things\do an cuoi ki mon may hoc\hand_Written(CNN)\merged_data"
os.makedirs(dest_dir, exist_ok=True)

# Bộ đếm để đặt tên file
counter = 1

# Duyệt qua tất cả các thư mục và file
for root, dirs, files in os.walk(source_dir):
    for file in tqdm(files, desc="Đang xử lý"):
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            # Lấy đường dẫn đầy đủ của file nguồn
            src_path = os.path.join(root, file)
            
            # Kiểm tra xem file có phải là ảnh chữ số viết tay không
            # (Giả sử tên file có dạng "số_mssv..." như 0_2152000...)
            try:
                # Tách số đầu tiên trong tên file (chữ số được nhận diện)
                digit = file.split('_')[0]
                if digit.isdigit() and 0 <= int(digit) <= 9:
                    # Tạo tên file mới
                    new_name = f"{digit}_{counter:06d}{os.path.splitext(file)[1]}"
                    dest_path = os.path.join(dest_dir, new_name)
                    
                    # Copy file sang thư mục đích
                    shutil.copy2(src_path, dest_path)
                    counter += 1
            except:
                continue

print(f"Đã gộp xong {counter-1} ảnh vào thư mục {dest_dir}")
```

# Xử lí ảnh

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from torchvision.transforms import v2

# ----------------------------
# 1. Tiền xử lý dữ liệu
# ----------------------------
class DigitDataset(Dataset):
    def __init__(self, image_paths, transform=None):
        self.image_paths = image_paths
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        
        # Đọc ảnh và tiền xử lý
        image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Đọc ảnh xám
        image = cv2.resize(image, (28, 28))                # Resize về 28x28
        image = cv2.bitwise_not(image)                     # Đảo màu (nền đen, chữ trắng)
        
        # Lấy label
        label = int(os.path.basename(img_path).split('_')[0])
        
        # Áp dụng transform
        if self.transform:
            image = self.transform(image)
            
        return image, label

# ----------------------------
# 2. Xây dựng mô hình CNN
# ----------------------------
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.conv2 = nn.Sequential(
            nn.Conv2d(32, 64, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 7 * 7, 128),
            nn.ReLU(),
            nn.Linear(128, 10)
        )

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.fc(x)
        return x

# ----------------------------
# 3. Hàm huấn luyện và đánh giá
# ----------------------------
def train(model, train_loader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    correct = 0
    
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        correct += (predicted == labels).sum().item()
    
    accuracy = 100 * correct / len(train_loader.dataset)
    return total_loss / len(train_loader), accuracy

def evaluate(model, test_loader, device):
    model.eval()
    correct = 0
    
    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            correct += (predicted == labels).sum().item()
    
    accuracy = 100 * correct / len(test_loader.dataset)
    return accuracy

# ----------------------------
# 4. Chuẩn bị dữ liệu
# ----------------------------
def load_data(data_dir):
    image_paths = []
    for root, _, files in os.walk(data_dir):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                image_paths.append(os.path.join(root, file))
    
    # Chia tập train/test (80/20)
    train_paths, test_paths = train_test_split(image_paths, test_size=0.2, random_state=1)
    
    # Transform
    transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))
    ])
    
    train_dataset = DigitDataset(train_paths, transform)
    test_dataset = DigitDataset(test_paths, transform)
    
    train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=64)
    
    return train_loader, test_loader

# ----------------------------
# 5. Hàm main
# ----------------------------
def main():
    # Thiết lập device
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")
    
    # Load dữ liệu
    data_dir = r"D:\code_things\do an cuoi ki mon may hoc\hand_Written(CNN)\merged_data"  
    train_loader, test_loader = load_data(data_dir)
    
    # Khởi tạo mô hình
    model = CNN().to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    
    # Huấn luyện
    epochs = 25
    best_acc = 0
    
    for epoch in range(epochs):
        train_loss, train_acc = train(model, train_loader, criterion, optimizer, device)
        test_acc = evaluate(model, test_loader, device)
        
        print(f"Epoch {epoch+1}/{epochs}:")
        print(f"  Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%")
        print(f"  Test Acc: {test_acc:.2f}%")
        
        # Lưu mô hình tốt nhất
        if test_acc > best_acc:
            best_acc = test_acc
            torch.save(model.state_dict(), 'best_model.pth')
    
    print(f"Best Test Accuracy: {best_acc:.2f}%")

if __name__ == "__main__":
    main()
```

# Du doan

```python
import os
import cv2
import torch
import pandas as pd
from torchvision import transforms

# ------------
# 1. Load mô hình đã huấn luyện
# ------------
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        # Tầng tích chập 1
        self.conv1 = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        # Tầng tích chập 2
        self.conv2 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        # Tầng fully connected
        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 7 * 7, 128),  # 64 channels * 7x7 size sau maxpool
            nn.ReLU(),
            nn.Linear(128, 10)  # 10 lớp đầu ra (0-9)
        )

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.fc(x)
        return x
    
    
def load_model(model_path='best_model.pth'):
    model = CNN()
    model.load_state_dict(torch.load(model_path))
    model.eval()
    return model

# ------------
# 2. Hàm tiền xử lý ảnh
# ------------
def preprocess_image(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        return None
    image = cv2.resize(image, (28, 28))
    image = cv2.bitwise_not(image)  # Đảo màu (nền đen, chữ trắng)
    image = transforms.ToTensor()(image)
    image = transforms.Normalize((0.5,), (0.5,))(image)
    return image.unsqueeze(0)  # Thêm chiều batch

# ------------
# 3. Dự đoán và xuất CSV
# ------------
def predict_and_export(input_dir, output_csv='predictions.csv'):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = load_model().to(device)
    
    results = []
    
    for filename in os.listdir(input_dir):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(input_dir, filename)
            
            # Tiền xử lý ảnh
            image = preprocess_image(image_path)
            if image is None:
                continue

            # Chuyển ảnh sang GPU nếu có
            image = image.to(device)
            
            # Dự đoán
            with torch.no_grad():
                output = model(image)
                _, predicted = torch.max(output, 1)
                digit = predicted.item()
            
            results.append([filename, digit])
    
    # Xuất kết quả ra CSV
    df = pd.DataFrame(results, nscolum=['filename', 'predicted_digit'])
    df.to_csv(output_csv, index=False)
    print(f"Đã xuất kết quả dự đoán vào file: {output_csv}")
# ------------
# 4. Chạy chương trình
# ------------
if __name__ == "__main__":
    input_dir = "D:\code_things\do an cuoi ki mon may hoc\hand_Written(CNN)\data no label\data.2025" 
    predict_and_export(input_dir)
```
